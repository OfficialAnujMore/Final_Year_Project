```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import time
import matplotlib.pyplot as plt 
import random
import csv 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
```

```python
pd.set_option('display.max_rows',300)
pd.set_option('display.max_columns',100)
pd.set_option('max_colwidth',800)

plt.figure(figsize=(20,10))
```

```python
df = pd.read_csv('model_dataset.csv')
```

```python
df.drop('Sr.No', axis = 'columns', inplace = True)
```

# Initial Dataframe

```python
df
```

# Shuffling the whole dataset

```python
df = shuffle(df)
df
```

# Obtaining dummies for JOB_NAME

```python
dummies = pd.get_dummies(df.Job_Name)
dummies
```

# Concating both datasets

```python
df1 = pd.concat([dummies, df], axis = 'columns')
df1
```

# Droping Job_Name

```python
df2 = df1.drop('Job_Name', axis = 'columns')
df2.sample(frac=1)
df2
```

# Generating X and y values

```python
# To generate x we drop salary
X = df2.drop('Salary', axis = 'columns')
X
```

```python
# To obtain y we consider only salary
y = df2.Salary
y
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size = 0.2, random_state=10)
```

```python
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)*100
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10,  test_size=0.2, random_state =0)

cvs = cross_val_score(LinearRegression(), X,y, cv=cv)
cvs
```

```python
cvs.mean()
```
