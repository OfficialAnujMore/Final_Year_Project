# Generating X and y values

```python
df = pd.read_csv('my_csv_shuffle.csv')
df
X = df.drop('Salary', axis = 'columns')
X
y = df.Salary
y
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size = 0.2, random_state=10)
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)*100
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10,  test_size=0.2, random_state =0)

cvs = cross_val_score(LinearRegression(), X,y, cv=cv)
cvs
```

```python

new_output=[]
for i in range(0,9):
    new_output.append(1)
for i in range(9,66):
    new_output.append(0)
print(new_output)
output=lr_clf.predict([new_output])
print(output) 
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size = 0.2)
regressor = RandomForestRegressor(n_estimators = 100, random_state = 10)
regressor.fit(X_train, y_train)
regressor.score(X_test,y_test)*100
```
